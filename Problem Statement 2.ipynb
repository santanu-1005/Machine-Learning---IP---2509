```python
df['Gender'].fillna(method='ffill', inplace=True)
df['City'].fillna(method='ffill', inplace=True)
df['New College Name'].fillna(method='ffill', inplace=True)
df['Colleges'].fillna(method='ffill', inplace=True)
df['Academic Year'].fillna(method='ffill', inplace=True)
df['Branch/ Specialisation'].fillna(method='ffill', inplace=True)
```

```python
df.dropna(
    subset=[
        'Gender', 
        'City', 
        'Position', 
        'New College Name', 
        'Colleges', 
        'Academic Year', 
        'Branch/ Specialisation', 
        'Other Branch', 
        'What is your current academic year?', 
        'Company Name/ College Name',
        'How did you come to know about this event?'
    ],
    how='all',
    inplace=True
)
df.reset_index(drop=True, inplace=True)

df
```
